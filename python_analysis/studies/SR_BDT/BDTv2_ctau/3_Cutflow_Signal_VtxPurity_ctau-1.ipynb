```python
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
from coffea import util
from coffea.processor import accumulate
import os
import re
import pandas as pd
import hist
import matplotlib as mpl
import mplhep

import utils
```

# All Lxy Bin

```python
prefix = './coffea/skimmed/'
suffix = '_skimmed'
```

```python
# Signal
sig_histo = util.load(f'{prefix}/ctau-1-BDT-based_AllLxy.coffea')[0]
```

```python
si = utils.get_signal_point_dict(sig_histo)

pd.options.display.float_format = '{:,.1f}'.format
si
```

```python
cutnames = utils.get_signal_list_of_cuts(sig_histo)

cutnames
```

### Vertex purity = n(Events where the selected vertex is gen-matched) / n(Events)

```python
pd.options.display.float_format = '{:,.2f}'.format
```

```python
eff_cutflow = utils.get_signal_cutflow_dict(sig_histo, 'cutflow_vtx_matched', do_only_gen_ee_reconstructed = False)
```

```python
table = pd.DataFrame.from_dict(eff_cutflow, orient='index')
table.columns = cutnames
```

```python
ct = 1.0

points = (si["ctau"] == ct)

df = table.loc[list(si[points].name)]
df = df.drop(columns='Preselections')  # dummy value stored at this cut step
df
```
